```python
from tkinter import * # Import the essential modules for the graphical user interface.
from tkinter import ttk
import tkinter.messagebox # For MessageBox widget.
rt = Tk() # The root.

rt.title("RPN") # The title of the application.
rt.geometry("546x596") # The window should not be resizable.
mf = ttk.Frame(rt)
argument, stacklist = StringVar(), StringVar() # For changing the content of the certain elements of the application.
mf.grid(column = 0, row = 0, sticky = (N, W, E, S)) # Everything is displayed in a grid.
label0 = Label(mf, textvariable = stacklist,foreground="black", background="white",width=54,font=("Arial Bold",12),height=5)
label0.grid(row = 0, columnspan = 4)
label1 = Label(mf, textvariable = argument, height=6,width=19,foreground="white", background="black")
label1.grid(row = 0, column = 3)
stack, command, num = [], '', ''
global history #this one contains claculator history (It's cleared after you exit)
history = ['None']
# The functions for "constructing" the number that is to be added to the stack.
def construct0():
    global num, argument
    if num == '0':
        tkinter.messagebox.showinfo("The error", "Can't begin with a zero.")
    else:
        num += '0'
        argument.set(num)
def construct1():
    global num, argument
    num += '1'
    argument.set(num)
def construct2():
    global num, argument
    num += '2'
    argument.set(num)
def construct3():
    global num, argument
    num += '3'
    argument.set(num)
def construct4():
    global num, argument
    num += '4'
    argument.set(num)
def construct5():
    global num, argument
    num += '5'
    argument.set(num)
def construct6():
    global num, argument
    num += '6'
    argument.set(num)
def construct7():
    global num, argument
    num += '7'
    argument.set(num)
def construct8():
    global num, argument
    num += '8'
    argument.set(num)
def construct9():
    global num, argument
    num += '9'
    argument.set(num)
def constructd(): # The dot.
    global num, argument
    if '.' in num:
        tkinter.messagebox.showinfo("The error", "A dot is used already.")
    elif num == '':
        num += "0."
        argument.set(num)
    else:
        num += '.'
        argument.set(num)
def constructm(): # The minus.
    global num, argument
    if num == '':
        num += '-'
        argument.set(num)
    else:
        tkinter.messagebox.showinfo("The error", "A minus can be put only in the beginning.")
def goback(): # The "backspace".
    global num, argument
    if num == '':
        tkinter.messagebox.showinfo("The error", "It's empty already.")
    else:
        num = num[:-1]
        argument.set(num)
def delfromstack(): # Deleting elements from the stack.
    global stacklist, stack
    if len(stack) == 0:
        tkinter.messagebox.showinfo("The error", "There is nothing on the stack.")
    else:
        stack = stack[:-1]
        if len(stack) == 0:
            stacklist.set('')
        else:
            stacklist.set(', '.join([str(x) for x in stack]))
def push(): # Pushing on the stack.
    global num, stacklist, argument
    if float(num) % 1 == 0.0:
        stack.append(int(num))
    else:
        stack.append(float(num))
    stacklist.set(', '.join([str(x) for x in stack]))
    argument.set('')
    num = ''
def factorial(number):
    if number == 0:
        return 1
    else:
        return number * factorial(number - 1)
def performa(): # Addition.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    else:
        stack[-2] = stack[-2] + stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performs(): # Substraction.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    else:
        stack[-2] = stack[-2] - stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performm(): # Multiplication.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    else:
        stack[-2] = stack[-2] * stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performd(): # Division.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack ais required.")
    else:
        if stack[-1] == 0:
            tkinter.messagebox.showinfo("The error", "The divisior is 0, can't perform the division.")
        else:
            stack[-2] = int((stack[-2] / stack[-1]) * 1000000 + 0.5) / 1000000.0
            del stack[-1]
            if stack[-1] % 1 == 0.0:
                stack[-1] = int(stack[-1])
            stacklist.set(', '.join([str(x) for x in stack]))
def performf(): # Factorial.
    global stack
    if len(stack) < 1:
        tkinter.messagebox.showinfo("The error", "At least 1 element in the stack is required.")
    elif type(stack[-1]) != int:
        tkinter.messagebox.showinfo("The error", "The type of the argument is not an integer.")
    elif stack[-1] < 0:
        tkinter.messagebox.showinfo("The error", "Unable to factor a negative number.")
    elif stack[-1] > 25:
        tkinter.messagebox.showinfo("The error", "The number is too large.")
    else:
        stack[-1] = factorial(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performp(): # The power.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    else:
        stack[-2] = stack[-2] ** stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
def performr(): # The root.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    else:
        stack[-2] = stack[-2] ** stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
def performmd(): # The modulo.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("The error", "At least 2 elements in the stack are required.")
    elif stack[-1] >= 0:
        stack[-2] = stack[-2] % stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
    else:
        stack[-1] = -stack[-1]
        stack[-2] = stack[-2] % stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
        
def hoverbutton(b):
    b.config(bg = 'white')

def hoverbuttonleave(b, c):
    b.config(bg = c)


            # Drawing the grid.
#starting screen

rt.columnconfigure(0, weight=1)
rt.rowconfigure(0, weight=1)        
bt0 = Button(mf, text = "7",bg = "light green", fg = 'Black', font = ('airal', 30), command = construct7)
bt1 = Button(mf, text = "8", command = construct8,  bg = "light green", fg = 'Black', font = ('airal', 30))
bt2 = Button(mf, text = "9", command = construct9, bg = "light green", fg = 'Black', font = ('airal', 30))
bt3 = Button(mf, text = "+", command = performa,   bg = "light green", fg = 'Black', font = ('airal', 30))
bt4 = Button(mf, text = "4", command = construct4,  bg = "light green", fg = 'Black', font = ('airal', 30))
bt5 = Button(mf, text = "5", command = construct5, bg = "light green", fg = 'Black', font = ('airal', 30))
bt6 = Button(mf, text = "6", command = construct6, bg = "light green", fg = 'Black', font = ('airal', 30))
bt7 = Button(mf, text = "-", command = performs,  bg = "light green", fg = 'Black', font = ('airal', 30))
bt8 = Button(mf, text = "1", command = construct1,  bg = "light green", fg = 'Black', font = ('airal', 30))
bt9 = Button(mf, text = "2", command = construct2,  width = 18, height = 6, background="black",fg="white")
bt10 = Button(mf, text = "3", command = construct3,  width = 18, height = 6, background="black",fg="white")
bt11 = Button(mf, text = "*", command = performm,  width = 18, height = 6, background="black",fg="white")
bt12 = Button(mf, text = "0", command = construct0, width = 18, height = 6, background="black",fg="white")
bt13 = Button(mf, text = ".", command = constructd, width = 18, height = 6, background="black",fg="white")
bt14 = Button(mf, text = "!", command = performf, width = 18, height = 6, background="black",fg="white")
bt15 = Button(mf, text = "/", command = performd, width = 18, height = 6, background="black",fg="white")
bt16 = Button(mf, text = "Push!", command = push, width = 18, height = 6, background="black",fg="white")
bt17 = Button(mf, text = "neg", command = constructm, width = 18, height = 6, background="black",fg="white")
bt18 = Button(mf, text = "<-", command = goback,  width = 18, height = 6, background="black",fg="white")
bt19 = Button(mf, text = "<- (st)", command = delfromstack, width = 18, height = 6, background="black",fg="white")

bt0.grid(column = 0, row = 2)
bt1.grid(column = 1, row = 2)
bt2.grid(column = 2, row = 2)
bt3.grid(column = 3, row = 2)
bt4.grid(column = 0, row = 3)
bt5.grid(column = 1, row = 3)
bt6.grid(column = 2, row = 3)
bt7.grid(column = 3, row = 3)
bt8.grid(column = 0, row = 4)
bt9.grid(column = 1, row = 4)
bt10.grid(column = 2, row = 4)
bt11.grid(column = 3, row = 4)
bt12.grid(column = 0, row = 5)
bt13.grid(column = 1, row = 5)
bt14.grid(column = 2, row = 5)
bt15.grid(column = 3, row = 5)
bt16.grid(column = 3, row = 1)
bt17.grid(column = 2, row = 1)
bt18.grid(column = 1, row = 1)
bt19.grid(column = 0, row = 1)
#cool button light up stuff for the standard calc
bt1.bind("<Enter>", lambda e: hoverbutton(bt1))
bt1.bind("<Leave>", lambda e: hoverbuttonleave(bt1, 'light blue'))
bt2.bind("<Enter>", lambda e: hoverbutton(bt2))
bt2.bind("<Leave>", lambda e: hoverbuttonleave(bt2, 'light blue'))
bt3.bind("<Enter>", lambda e: hoverbutton(bt3))
bt3.bind("<Leave>", lambda e: hoverbuttonleave(bt3, 'light blue'))
bt4.bind("<Enter>", lambda e: hoverbutton(bt4))
bt4.bind("<Leave>", lambda e: hoverbuttonleave(bt4, 'light blue'))
bt5.bind("<Enter>", lambda e: hoverbutton(bt5))
bt5.bind("<Leave>", lambda e: hoverbuttonleave(bt5, 'light blue'))
bt6.bind("<Enter>", lambda e: hoverbutton(bt6))
bt6.bind("<Leave>", lambda e: hoverbuttonleave(bt6, 'light blue'))
bt7.bind("<Enter>", lambda e: hoverbutton(bt7))
bt7.bind("<Leave>", lambda e: hoverbuttonleave(bt7, 'light blue'))
bt8.bind("<Enter>", lambda e: hoverbutton(bt8))
bt8.bind("<Leave>", lambda e: hoverbuttonleave(bt8, 'light blue'))
bt9.bind("<Enter>", lambda e: hoverbutton(bt9))
bt9.bind("<Leave>", lambda e: hoverbuttonleave(bt9, 'light blue'))
bt0.bind("<Enter>", lambda e: hoverbutton(bt0))
bt0.bind("<Leave>", lambda e: hoverbuttonleave(bt0, 'light blue'))
bt10.bind("<Enter>", lambda e: hoverbutton(bt10))
bt10.bind("<Leave>", lambda e: hoverbuttonleave(bt10, 'light blue'))
bt11.bind("<Enter>", lambda e: hoverbutton(bt11))
bt11.bind("<Leave>", lambda e: hoverbuttonleave(bt11, 'light blue'))
bt12.bind("<Enter>", lambda e: hoverbutton(bt12))
bt12.bind("<Leave>", lambda e: hoverbuttonleave(bt12, 'light blue'))
bt13.bind("<Enter>", lambda e: hoverbutton(bt13))
bt13.bind("<Leave>", lambda e: hoverbuttonleave(bt13, 'light blue'))
bt14.bind("<Enter>", lambda e: hoverbutton(bt4))
bt14.bind("<Leave>", lambda e: hoverbuttonleave(bt14, 'light blue'))
bt15.bind("<Enter>", lambda e: hoverbutton(bt15))
bt15.bind("<Leave>", lambda e: hoverbuttonleave(bt15, 'light blue'))
bt16.bind("<Enter>", lambda e: hoverbutton(bt16))
bt16.bind("<Leave>", lambda e: hoverbuttonleave(bt16, 'light blue'))
bt17.bind("<Enter>", lambda e: hoverbutton(bt17))
bt17.bind("<Leave>", lambda e: hoverbuttonleave(bt17, 'light blue'))
bt18.bind("<Enter>", lambda e: hoverbutton(bt18))
bt18.bind("<Leave>", lambda e: hoverbuttonleave(bt18, 'light blue'))
bt19.bind("<Enter>", lambda e: hoverbutton(bt19))
bt19.bind("<Leave>", lambda e: hoverbuttonleave(bt19, 'light blue'))



rt.mainloop() # The main loop of the application in order to make it more IDLE-friendly.
```
