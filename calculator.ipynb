```python
from tkinter import * 
from tkinter import ttk
import tkinter.messagebox
rt = Tk() 
rt.title("Calculatrice polonaise") 
rt.resizable(0, 0)
mf = ttk.Frame(rt)
argument, stacklist = StringVar(), StringVar()
mf.grid(column = 0, row = 0, sticky = (N, W, E, S))
label0 = Label(mf, textvariable = stacklist)
label0.grid(row = 0, columnspan = 3)
label1 = Label(mf, textvariable = argument, background = "white",width=17,height=5)
label1.grid(row = 0, column = 3)
stack, command, num = [], '', ''


def construct0():
    global num, argument
    if num == '0':
        tkinter.messagebox.showinfo("Impossible de commencer avec un zéro")
    else:
        num += '0'
        argument.set(num)
def construct1():
    global num, argument
    num += '1'
    argument.set(num)
def construct2():
    global num, argument
    num += '2'
    argument.set(num)
def construct3():
    global num, argument
    num += '3'
    argument.set(num)
def construct4():
    global num, argument
    num += '4'
    argument.set(num)
def construct5():
    global num, argument
    num += '5'
    argument.set(num)
def construct6():
    global num, argument
    num += '6'
    argument.set(num)
def construct7():
    global num, argument
    num += '7'
    argument.set(num)
def construct8():
    global num, argument
    num += '8'
    argument.set(num)
def construct9():
    global num, argument
    num += '9'
    argument.set(num)
def constructd(): 
    global num, argument
    if '.' in num:
        tkinter.messagebox.showinfo("Un point est déjà utilisé")
    elif num == '':
        num += "0."
        argument.set(num)
    else:
        num += '.'
        argument.set(num)
def constructm(): 
    global num, argument
    if num == '':
        num += '-'
        argument.set(num)
    else:
        tkinter.messagebox.showinfo("Un moins ne peut être mis qu’au début")
def goback(): 
    global num, argument
    if num == '':
        tkinter.messagebox.showinfo("déjà vide")
    else:
        num = num[:-1]
        argument.set(num)
def delfromstack():
    global stacklist, stack
    if len(stack) == 0:
        tkinter.messagebox.showinfo("Il n’y a rien dans la pile")
    else:
        stack = stack[:-1]
        if len(stack) == 0:
            stacklist.set('')
        else:
            stacklist.set(', '.join([str(x) for x in stack]))
def push(): 
    global num, stacklist, argument
    if float(num) % 1 == 0.0:
        stack.append(int(num))
    else:
        stack.append(float(num))
    stacklist.set(', '.join([str(x) for x in stack]))
    argument.set('')
    num = ''
def factorial(number):
    if number == 0:
        return 1
    else:
        return number * factorial(number - 1)
def performa(): # Addition.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires")
    else:
        stack[-2] = stack[-2] + stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performs(): # Substraction.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    else:
        stack[-2] = stack[-2] - stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performm(): # Multiplication.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    else:
        stack[-2] = stack[-2] * stack[-1]
        del stack[-1]
        if stack[-1] % 1 == 0.0:
            stack[-1] = int(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performd(): # Division.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    else:
        if stack[-1] == 0:
            tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
        else:
            stack[-2] = int((stack[-2] / stack[-1]) * 1000000 + 0.5) / 1000000.0
            del stack[-1]
            if stack[-1] % 1 == 0.0:
                stack[-1] = int(stack[-1])
            stacklist.set(', '.join([str(x) for x in stack]))
def performf(): # Factorial.
    global stack
    if len(stack) < 1:
        tkinter.messagebox.showinfo("Au moins 1 éléments dans la pile est nécessaires.")
    elif type(stack[-1]) != int:
        tkinter.messagebox.showinfo("Le type de l’argument n’est pas un entier.")
    elif stack[-1] < 0:
        tkinter.messagebox.showinfo("Impossible de prendre en compte un nombre négatif.")
    elif stack[-1] > 25:
        tkinter.messagebox.showinfo("Le nombre est trop grand.")
    else:
        stack[-1] = factorial(stack[-1])
        stacklist.set(', '.join([str(x) for x in stack]))
def performp(): # power.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    else:
        stack[-2] = stack[-2] ** stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
def performr(): # root.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    else:
        stack[-2] = stack[-2] ** stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
def performmd(): # modulo.
    global stack
    if len(stack) < 2:
        tkinter.messagebox.showinfo("Au moins 2 éléments dans la pile sont nécessaires.")
    elif stack[-1] >= 0:
        stack[-2] = stack[-2] % stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))
    else:
        stack[-1] = -stack[-1]
        stack[-2] = stack[-2] % stack[-1]
        del stack[-1]
        stacklist.set(', '.join([str(x) for x in stack]))

# Traçant la grille.
bt0 = Button(mf, text = "7", command = construct7, width = 17, height = 7)
bt1 = Button(mf, text = "8", command = construct8, width = 17, height = 7)
bt2 = Button(mf, text = "9", command = construct9, width = 17, height = 7)
bt3 = Button(mf, text = "+", command = performa, width = 17, height = 7)
bt4 = Button(mf, text = "4", command = construct4, width = 17, height = 7)
bt5 = Button(mf, text = "5", command = construct5, width = 17, height = 7)
bt6 = Button(mf, text = "6", command = construct6, width = 17, height = 7)
bt7 = Button(mf, text = "-", command = performs, width = 17, height = 7)
bt8 = Button(mf, text = "1", command = construct1, width = 17, height = 7)
bt9 = Button(mf, text = "2", command = construct2, width = 17, height = 7)
bt10 = Button(mf, text = "3", command = construct3, width = 17, height = 7)
bt11 = Button(mf, text = "*", command = performm, width = 17, height = 7)
bt12 = Button(mf, text = "0", command = construct0, width = 17, height = 7)
bt13 = Button(mf, text = ".", command = constructd, width = 17, height = 7)
bt14 = Button(mf, text = "!", command = performf, width = 17, height = 7)
bt15 = Button(mf, text = "/", command = performd, width = 17, height = 7)
bt16 = Button(mf, text = "Push!", command = push, width = 17, height = 7)
bt17 = Button(mf, text = "neg", command = constructm, width = 17, height = 7)
bt18 = Button(mf, text = "<-", command = goback, width = 17, height = 7)
bt19 = Button(mf, text = "<- (st)", command = delfromstack, width = 17, height = 7)


#
bt20 = Button(mf, text = "A", command = push, width = 17, height = 7)
bt21 = Button(mf, text = "B", command = constructm, width = 17, height = 7)
bt22 = Button(mf, text = "C", command = goback, width = 17, height = 7)
bt23 = Button(mf, text = "D", command = delfromstack, width = 17, height = 7)
#

bt0.grid(column = 0, row = 2)
bt1.grid(column = 1, row = 2)
bt2.grid(column = 2, row = 2)
bt3.grid(column = 3, row = 2)
bt4.grid(column = 0, row = 3)
bt5.grid(column = 1, row = 3)
bt6.grid(column = 2, row = 3)
bt7.grid(column = 3, row = 3)
bt8.grid(column = 0, row = 4)
bt9.grid(column = 1, row = 4)
bt10.grid(column = 2, row = 4)
bt11.grid(column = 3, row = 4)
bt12.grid(column = 0, row = 5)
bt13.grid(column = 1, row = 5)
bt14.grid(column = 2, row = 5)
bt15.grid(column = 3, row = 5)
bt16.grid(column = 0, row = 6)
bt17.grid(column = 1, row = 6)
bt18.grid(column = 2, row = 6)
bt19.grid(column = 3, row = 6)

#
bt20.grid(column = 3, row = 7)
bt21.grid(column = 2, row = 7)
bt22.grid(column = 1, row = 7)
bt23.grid(column = 0, row = 7)
#
rt.geometry('520x575')
rt.minsize(470, 770)
rt.mainloop() # main loop
```
